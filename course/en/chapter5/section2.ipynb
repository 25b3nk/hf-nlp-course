```python
# prompt: split text column into two using tab as delimiter

from datasets import load_dataset
import pandas as pd

# Load the dataset
dataset = load_dataset("text", data_files="/content/SMSSpamCollection")

# Convert the dataset to a pandas DataFrame
df = pd.DataFrame(dataset['train'])

# Split the text column based on the tab delimiter
df[['label', 'text']] = df['text'].str.split('\t', expand=True)

# Display the updated DataFrame
print(df.head())
```
